###On average, how much money each company has received at Series A round?

```python
#On average, how much money each company has received at Series A round?
cols_keep=['Series A']
df_investment4=df_investment[df_investment['FUNDING_TYPE'].isin(cols_keep)]
df_investment4['MONEY_RAISED1'].replace('NaN',int(0))
df_investment4.groupby(['COMPANY_NAME','FUNDING_TYPE'])['MONEY_RAISED1'].sum().to_frame().tail()
```
